```
N=130
n=[]
n=[9,39,40,24,11,7]
sum1=0
for i in range(6):
  sum1=sum1+n[i]
for i in range(6):
  n[i]=n[i]*i
sum2=0
for i in range(6):
  sum2=sum2+n[i]
L=(1/N)*sum2
print(round(L,3))
```
