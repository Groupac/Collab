# **Зачетная задача №1:**
Написать функцияю python для генерации случайной величины, имеющей экспоненциальное распределение с параметром lambda на основе равномерно распределенной случайной величины (np.random.uniform).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv('https://raw.githubusercontent.com/andrey-gl/study2020/master/data/regr2.csv')

x = df.iloc[:,2:]
y = df.iloc[:,1]
print(x)
print('Таблица y')
print(y)

plt.scatter(y,x['x1'])
plt.show()

plt.hist(x['x1'], bins = 20)
plt.show()
plt.hist(x['x2'], bins = 20)
plt.show()
plt.hist(x['x3'], bins = 20)
plt.show()

plt.scatter(y,x['x2'])
plt.show()


plt.scatter(y,x['x3'])
plt.show()

lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
print('Коэффициенты: ',lr.coef_)
print(lr.score(x,y))

print('Таблица squared_error:')
print(mean_squared_error(y,y_pred))
```
